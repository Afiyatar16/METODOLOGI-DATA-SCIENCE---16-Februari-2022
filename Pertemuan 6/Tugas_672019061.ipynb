## **Permen** - Pie Chart

## **Makanan** Bar Chart


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
datapenjualan_makananpenutup = ('Donat','Pastel','Kue Coklat','Ice Cream','Puding Vanila','Brownies','Puding Strawberi','Puding Coklat','Ice Cream Nutela','Kue Coklat-Keju','Kue Wajik','Kue Sus','Mochi')
banyak = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({'penjualan':datapenjualan_makananpenutup, 'banyakshort':banyak})
df.sort_values(by='banyakshort', inplace = True,ascending=False)

kor_x = np.arange(len(df))
warna = ['#1976D2' for _ in range(len(df))]
warna [-3] = '#FF0000'
warna [-2] = '#FF0000'
warna [-1] = '#FF0000'
plt.bar(kor_x, df['banyakshort'],tick_label=df['penjualan'],color = warna)
plt.xticks(rotation=90)
plt.ylabel('Populasi (juta)')
plt.title('Populasi 10 Negara di Dunia')
plt.xlabel('Nama Negara')
plt.show()
```

Tiga makanan penutup yang akan dihapus karena sedikit peminat adalah : Pudding Vanilla, Pastel, Kue Wajik.

Alasan menggunakan Bar Chart adalah mempermudah melihat grafik penjualan makanan sehingga bisa diurutkan baik dari yang terlaris sampai paling sedikit peminat.
