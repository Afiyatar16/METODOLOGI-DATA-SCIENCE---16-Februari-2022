```python
#imputasi modus
from sklearn.impute import SimpleImputer
mobil = {'mobil':['Ford','Ford','Honda','Tesla',np.nan,
         'Mitsubhisi',np.nan,'Tosa']}
dataM = pd.DataFrame(mobil)
dataM
```

```python
imp = SimpleImputer(strategy='most_frequent')
imp.fit_transform(dataM)
```
